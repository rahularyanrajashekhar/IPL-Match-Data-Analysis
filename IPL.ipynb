```python
#Load the dataset and demonstrate the successful loading of the dataset by displaying both the top and bottom five rows.
import pandas as pd

try:
  filepath = '/content/drive/MyDrive/43031/Assessment_Task2/deliveries.csv'
  deliveries_dataset = pd.read_csv(filepath, header=0)
  print("Dataset successfully loaded.")

  print("Top 5 rows: ")
  print(deliveries_dataset.head()) #displays top 5 rows

  print("Bottom 5 rows: ")
  print(deliveries_dataset.tail()) #displays bottom 5 rows
except Exception as ex:
    print("An error occurred:", ex)
```

```python
#Size of the dataset
print("Side of the dataset: ")
print(deliveries_dataset.shape)

#displaying the columns of the dataset
print("Below shows the column values: ")
print(deliveries_dataset.columns.values)

#displaying the datatype of the columns
print("The datatype of the columns: ")
print(deliveries_dataset.dtypes)
```

```python
try:
  selected_columns = ['match_no', 'date', 'venue', 'batting_team', 'bowling_team',
                    'innings', 'over', 'striker', 'bowler', 'runs_of_bat', 'extras',
                    'wide', 'legbyes', 'byes', 'noballs']
  filtered_dt = deliveries_dataset[selected_columns]
  print(filtered_dt.columns.values)
except exception as ex:
  print("An error occured: ", ex)
```

```python
# Convert 'date' column to datetime
deliveries_dataset['date'] = pd.to_datetime(deliveries_dataset['date'])

# Check the data types
print(deliveries_dataset['date'].dtypes)
```

```python
#Summary statistics
print("Summary statistics: ")
print(deliveries_dataset.describe(include='all'))
```

```python
print("Duplicates: " ,deliveries_dataset.duplicated()) # check for duplicates

print("Null values: ", deliveries_dataset.isnull()) #check for null values

print("Null values present in: ", deliveries_dataset.isnull().sum())
```
